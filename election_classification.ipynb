```python
# create and fill the column for each poll's error
df['Poll Margin'] = (df['adjpoll_trump'] - df['adjpoll_clinton']).astype('float64')
df['Actual Margin'] = (df['Donald Actual'] - df['Hilary Actual']).astype('float64')
df['Margin Error'] = (df['Poll Margin'] - df['Actual Margin'])
df['Trump Outperformed'] = (df['Margin Error'] > 0).astype('bool')
df['Trump Won'] = (df['Actual Margin'] > 0).astype('bool')
df['Trump'] = pd.get_dummies(df["Trump Won"])[1]
df['Trump Predicted'] = (df['Poll Margin'] > 0).astype('bool')
df['Trump Poll'] = pd.get_dummies(df["Trump Predicted"])[1]
df['Poll v Outcome'] = 'Trump Correct'
df.loc[(df['Trump Poll'] == 0) & (df['Trump Won'] == 0), 'Poll v Outcome'] = 'Clinton Correct'
df.loc[(df['Trump Poll'] == 0) & (df['Trump Won'] == 1), 'Poll v Outcome'] = 'Trump beats Poll'
df.loc[(df['Trump Poll'] == 1) & (df['Trump Won'] == 0), 'Poll v Outcome'] = 'Clinton beats Poll'
df['Poll Correct'] = 'Poll Correct'
df.loc[(df['Poll v Outcome'] == 'Trump beats Poll') | (df['Poll v Outcome'] == 'Clinton beats Poll'), 'Poll Correct'] = 'Poll Incorrect'
df['A Grade'] = pd.get_dummies(df["grade"])[['A', 'A+', 'A-']].sum(axis=1)
df['B Grade'] = pd.get_dummies(df["grade"])[['B', 'B+', 'B-']].sum(axis=1)
df['C Grade'] = pd.get_dummies(df["grade"])[['C', 'C+', 'C-']].sum(axis=1)
df['Likely Voters'] = pd.get_dummies(df["population"])[['lv']].sum(axis=1)
df['Registered Voters'] = pd.get_dummies(df["population"])[['rv', 'v', 'a']].sum(axis=1)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import cross_val_predict
from sklearn.model_selection import StratifiedKFold
from sklearn import preprocessing

from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import BernoulliNB
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegressionCV
from sklearn.linear_model import RidgeClassifier
from sklearn.model_selection import GroupKFold
```

```python
# model list
models = []

models.append(("KNN", KNeighborsClassifier()))
models.append(("DT", DecisionTreeClassifier()))
models.append(("BNB", BernoulliNB()))
models.append(("GNB", GaussianNB()))
models.append(("LogR", LogisticRegressionCV()))
```

```python
# features
features = ['Poll Margin']
X = df[features]

# responses
y = df.Trump
groups = df['state']
```

```python
gkf = GroupKFold(n_splits=10)

names = []
scores = []


for name, model in models:
    model
    score = cross_val_score(model, X, y, cv=gkf.get_n_splits(groups=groups), scoring='accuracy').mean()
    names.append(name)
    scores.append(score)
    
kf_cross_val = pd.DataFrame({'Name' : names, 'Score' : scores })
print(kf_cross_val)
```

```python

df_2020_m['Trump Outperformed'] = (df_2020_m['Margin Error'] > 0).astype('bool')
df_2020_m['Trump Won'] = (df_2020_m['Actual Margin'] > 0).astype('bool')
df_2020_m['Trump_D'] = pd.get_dummies(df_2020_m["Trump Won"])[1]
df_2020_m['Trump Predicted'] = (df_2020_m['Poll Margin'] > 0).astype('bool')
df_2020_m['Trump Poll'] = pd.get_dummies(df_2020_m["Trump Predicted"])[1]
df_2020_m['Poll v Outcome'] = 'Trump Correct'
df_2020_m.loc[(df_2020_m['Trump Poll'] == 0) & (df_2020_m['Trump Won'] == 0), 'Poll v Outcome'] = 'Biden Correct'
df_2020_m.loc[(df_2020_m['Trump Poll'] == 0) & (df_2020_m['Trump Won'] == 1), 'Poll v Outcome'] = 'Trump beats Poll'
df_2020_m.loc[(df_2020_m['Trump Poll'] == 1) & (df_2020_m['Trump Won'] == 0), 'Poll v Outcome'] = 'Biden beats Poll'
df_2020_m['Poll Correct'] = 'Poll Correct'
df_2020_m.loc[(df_2020_m['Poll v Outcome'] == 'Trump beats Poll') | (df_2020_m['Poll v Outcome'] == 'Biden beats Poll'), 'Poll Correct'] = 'Poll Incorrect'
df_2020_m['A Grade'] = pd.get_dummies(df_2020_m["grade"])[['A', 'A+', 'A-']].sum(axis=1)
df_2020_m['B Grade'] = pd.get_dummies(df_2020_m["grade"])[['B', 'B+', 'B-']].sum(axis=1)
df_2020_m['C Grade'] = pd.get_dummies(df_2020_m["grade"])[['C', 'C+', 'C-']].sum(axis=1)
df_2020_m['Likely Voters'] = pd.get_dummies(df_2020_m["population"])[['lv']].sum(axis=1)
df_2020_m['Registered Voters'] = pd.get_dummies(df_2020_m["population"])[['rv', 'v', 'a']].sum(axis=1)
```
